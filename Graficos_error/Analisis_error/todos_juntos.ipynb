```python
from os import listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mypath = 'datos_crudos' #path with files to analyze
from matplotlib.ticker import AutoMinorLocator
files = [f for f in listdir(mypath) if isfile(join(mypath, f))] #get all file names of that path
from scipy.optimize import curve_fit
```

```python
for file_name in files:
    file_dir = mypath + '/' + str(file_name)
    data = np.genfromtxt(file_dir) #2 x bins
    plt.plot(data[0,:],data[1,:],'-o',label=str(file_name))
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$R_7$ (log)')
plt.ylabel(r'Error (log)')
#plt.savefig('total.pdf')
```
